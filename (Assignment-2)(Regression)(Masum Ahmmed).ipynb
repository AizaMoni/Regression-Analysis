```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
column_names = f1.columns
```

```python
column_names
```

```python
label_columns = ['store_size']
```

```python
label_columns
```

```python
lb_encoders = {}
```

```python
for i in label_columns:
    lb_encoders[i] = LabelEncoder()
    f1[i] = lb_encoders[i].fit_transform(f1[i])
```

```python
f1.head()
```

```python
one_hot_columns = ['division', 'item_type']
```

```python
ohcs = {}
```

```python
for i in one_hot_columns:
    ohcs[i] = OneHotEncoder(drop='first')
    tmp = ohcs[i].fit_transform(f1[[i]]) 
    f1[ohcs[i].categories_[0][1:]] = tmp.toarray()
```

```python
f1.head()
```

```python
f1.drop(['division', 'item_type'], axis=1, inplace=True)
```

```python
f1.head()
```

```python
X = f1.drop(['total_price'], axis=1)
y = f1['total_price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.10, random_state=123)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LRM = LinearRegression()
```

```python
LRM.fit(X_train, y_train);
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, confusion_matrix, r2_score
```

```python
def eval(model, X_train, y_train, X_test, y_test):
    print("The training score is,", model.score(X_train, y_train),end='\n')
    print("The testing score is,", model.score(X_test, y_test))
```

```python
def metric_score(y_test, ypred):
    print("The mean absolute error is: ", mean_absolute_error(y_test, ypred))
    print("The mean squared error is: ", mean_squared_error(y_test, ypred))
    print("The R2 score is: ", r2_score(y_test, ypred))
```

```python
eval(LRM, X_train, y_train, X_test, y_test)
```

```python
metric_score(y_test, LRM.predict(X_test))
```
